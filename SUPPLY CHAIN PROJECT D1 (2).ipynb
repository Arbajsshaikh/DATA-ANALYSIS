```python
#8.seperate Mumbai's data & Kolkata's data using loc , calculate count of different Transportation modes for that 
MUM=X.loc[X['Location']=='Mumbai']
KOL=X.loc[X['Location']=='Kolkata']
```

```python
M = MUM['Transportation modes'].value_counts().reset_index()
K = KOL['Transportation modes'].value_counts().reset_index()
```

```python
M
```

```python
K
```

```python
F=pd.merge(M,K,on='Transportation modes',how='outer').rename(columns={'count_x':'MUMBAI','count_y':'KOLKATA'})
F
```

```python
plt.figure(figsize=(15,6))
plt.bar(F['Transportation modes'],F['MUMBAI'],color='red',label='Mumbai',align='center',width=0.5)
plt.bar(F['Transportation modes'],F['KOLKATA'],color='green',label='Kolkata',align='edge',width=0.5)
plt.legend()
plt.show()
```

```python
#9.create a new column named as 'Shipment' based on Shipping carriers & Supplier name to get proper information from both
       #i.e.Carrier B +  Supplier 3 ------>Carrier B3
       #i.e.Carrier C +  Supplier 5 ------>Carrier C5
        
X['Shipment']=X['Shipping carriers']+X['Supplier name'].apply(lambda x:x[-1])
X
```

```python
#10.calculate total profit for all cities and plot using line plot
    #i.e. Selling amount= Price* Number of products sold
    #i.e. manufacturing amount=Number of products sold * Manufacturing costs
    #i.e  Transportation charge =5% of Manufacturing costs for each product
    #i.e.total profit = Selling amount - (manufacturing amount + charge )
```

```python
X['Selling amount']=X['Price']*X['Number of products sold']
X['Manufacturing amount']=X['Manufacturing costs']*X['Number of products sold']
X['Transportation charge']=X['Manufacturing costs']*0.05
X
```

```python
X['total profit']=X['Selling amount'] - (X['Manufacturing amount'] + X['Transportation charge'] )
X
```

```python
XF=X.groupby('Location')['total profit'].sum().reset_index()
XF
```
